## 1 Symmetric difference

Let A and B be sets. The set (A\B)∪(B\A) is called the symmetric difference of the two sets. Implement this operation and verify your results. Compare your results to the result of the command: `A.symmetric_difference(B)`

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# A - B = {1, 2};
# B - A = {5, 6};
# A | B = {1, 2, 5, 6};
C = (A - B) | (B - A)

D = A.symmetric_difference(B)

print("C: ", C)
print("D: ", D)
print("C == D ?: ", C == D)
````

## 2 Build a list

Use the following equation to build a list of. [u₀, u₁, u₂, ..., u₁₀₀₀], where h=0.001, and a=-0.5.  
Note that u₀ = e⁰, u₁ = eʰᵃ, u₂ = e²ʰᵃ  
uₙ₊₃ = uₙ₊₂ + ha ( (23/12) uₙ₊₂ - (4/3) uₙ₊₁ + (5/12) uₙ )

````python
import math

h = 0.001
a = -0.5

u = []
u.append(math.exp(0))
u.append(math.exp(h * a))
u.append(math.exp(2 * h * a))

# n: 0 ~ 997 (3 ~ 1000)
for n in range(998):
    u.append(u[n + 2] + h * a * ((23 / 12) * u[n + 2] - (4 / 3) * u[n + 1] + (5 / 12) * u[n]))

print(u)
# for n in range(1000):
#     print(" {0:.4f} ".format(u[n]))
````

## 4 how_sparse

write a decorator for the function `how_sparse` which clean input
matrix (represented by lists in a list) A by setting the elements that
are less than 1.0e-16 to zero

````python
A = [
    [1, 2, 3],
    [4, 1.0e-17, 6],
    [7, 0, 9]
]

def clean_matrix(func):
    def wrapper(A):
        cleaned_A = [[0 if x < 1.0e-16 else x for x in row] for row in A]
        return func(cleaned_A)
    return wrapper

@clean_matrix
def how_sparse(A):
    total_ele = len(A) * len(A[0])
    zero_ele = sum(row.count(0) for row in A)
    return zero_ele / total_ele if total_ele > 0 else 0

print(how_sparse(A))
````
